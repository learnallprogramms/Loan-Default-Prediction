```python
# Import necessary libraries
import numpy as np  # For numerical operations
import pandas as pd  # For data manipulation and analysis
import matplotlib.pyplot as plt  # For data visualization
import seaborn as sns  # For enhanced data visualization

# Suppress warnings
import warnings  # For managing warnings in the code
warnings.filterwarnings('ignore')  # Ignore all warnings
```

```python
# Read the CSV file into a pandas DataFrame
df2 = pd.read_csv('Data_without_Scaling_&_without_Outliers_Treatment.csv')

# Display the first few rows of the DataFrame to inspect the data
df2.head()
```

# Hypothesis Testing for Numerical Columns

```python
from scipy.stats import shapiro, levene, mannwhitneyu

# Perform the sampling
x = df2[df2['V32'] == 1]['V6'].sample(n=1500, random_state=1).reset_index(drop=True)
y = df2[df2['V32'] == 0]['V6'].sample(n=1500, random_state=1).reset_index(drop=True)

# Normality check for x
test_stats, p_value_x = shapiro(x)
print('p_value for x :', p_value_x)
if p_value_x <= 0.05:
    print('Data is Not Normal')
else:
    print('Data is Normal')

print()

# Normality check for y
test_stats, p_value_y = shapiro(y)
print('p_value for y :', p_value_y)
if p_value_y <= 0.05:
    print('Data is Not Normal')
else:
    print('Data is Normal')

print()

# Variance check
test_stats, p_value_var = levene(x, y)
print('variances p_value :', p_value_var)
if p_value_var <= 0.05:
    print('Variances are Not Equal')
else:
    print('Variances are Equal')

print()

# Mann-Whitney U test (non-parametric, two groups)
u_stat, p_value_mw = mannwhitneyu(x, y)
print('whitney score :', round(u_stat, 2))
print('P_value for whitney :', round(p_value_mw, 6))

# Step 4: Decision rule based on alpha (significance level)
alpha = 0.05
if p_value_mw < alpha:
    decision = 'Significant'  # Feature is significant
else:
    decision = 'Not Significant'  # Feature might not be significant

# Step 5: Draw the conclusion
conclusion = f"U statistic: {u_stat}, p-value: {p_value_mw}, Decision: {decision}"
print(conclusion)
```

# Hypothesis Testing for categorical Columns

```python
from scipy.stats import chi2_contingency
```

```python
# Create a contingency table
contingency_table = pd.crosstab(df2.V10, df2.V32)

# Perform Chi-Square test
chi2_stat, p_value_ch, dof, ex = chi2_contingency(contingency_table)

print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-Value: {p_value_ch}")
print()

# Step 4: Decision rule based on alpha (significance level)
alpha = 0.05
if p_value_ch < alpha:
    decision = 'Significant'  # Feature is significant
else:
    decision = 'Not Significant'  # Feature might not be significant

# Step 5: Draw the conclusion
conclusion = f"Chi2 statistic: {chi2_stat}, p-value: {p_value_ch}, Decision: {decision}"
print(conclusion)
```
